```python
import os
import glob
import random
import numpy as np
import pandas as pd
from PIL import Image

import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

### Hindi/Devanagari MNIST

```python
DATA_FOLDER = "../data"

TRAIN_FOLDER_NAME = "Train"
TEST_FOLDER_NAME = "Test"

DEVANAGARI_DATA_FOLDER = os.path.join(os.path.join(DATA_FOLDER, "DevanagariHandwrittenCharacterDataset"))
print(DEVANAGARI_DATA_FOLDER)

# Data parameters
IMG_HEIGHT = 32
IMG_WIDTH = 32

labels_to_keep = [
    "digit_0", "digit_1", "digit_2", "digit_3", "digit_4", "digit_5", "digit_6", "digit_7", "digit_8", "digit_9"
]
num_classes = len(labels_to_keep)
class_names_to_labels = dict([(cls_name, lbl) for cls_name, lbl in zip(labels_to_keep, list(range(len(labels_to_keep))))])
# print(class_names_to_labels)
labels_to_class_names = dict([(v, k) for k, v in class_names_to_labels.items()])
```

```python
train_folders = glob.glob(os.path.join(DEVANAGARI_DATA_FOLDER, TRAIN_FOLDER_NAME, "*"))

images = []

for folder in train_folders:
    files = glob.glob(os.path.join(folder, "*"))
    set_of_files = random.sample(files, 20)
    # print(set_of_files)
    for f in set_of_files:
        im = Image.open(f)
        images.append(im)

random.shuffle(images)

columns = len(images) // 5
rows = len(images) // columns

fig = plt.figure(figsize=(100, 20))
for i, im in enumerate(images):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(im, cmap='gray', vmin=0, vmax=255)
    plt.axis("off")

plt.show()
```

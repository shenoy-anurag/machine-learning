## Random Data Generation
`generate_increasing_data_points` generates points which trend upwards, although they may go up and down randomly.

`generate_decreasing_data_points` generates points which trend downwards, although they may go up and down randomly.

`generate_erratic_data_points` generates points which have peaks and troughs, but don't trend either upward or downwards and instead stagnate.

```python
from random import randint


def generate_increasing_data_points(high, low, n_points):
    data_range = high - low
    step = data_range / n_points
    x = []
    for i in range(n_points):
        if i == 0:
            x.append(low)
        else:
            val = step * i * (1.0 + randint(-15, 20) * 0.01)
            x.append(val)
    return x


def generate_decreasing_data_points(high, low, n_points):
    data_range = high - low
    step = data_range / n_points
    x = []
    for i in range(n_points):
        if i == 0:
            x.append(high)
        else:
            val = step * i * (1.0 + randint(-15, 20) * 0.01)
            x.append(high - val)
    return x


def generate_erratic_data_points(high, low, n_points):
    data_range = high - low
    step = data_range / n_points
    x = []
    for i in range(n_points):
        if i == 0:
            x.append(data_range / 2)
        else:
            val = data_range / 2 * (1.0 + randint(-30, 50) * 0.01)
            x.append(val)
    return x
```

```python
num_points = 20
x = generate_increasing_data_points(high=500, low=30, n_points=num_points)
x2 = generate_decreasing_data_points(high=400, low=50, n_points=num_points)
x3 = generate_erratic_data_points(high=500, low=10, n_points=num_points)

plt.figure(figsize=(10, 7))
plt.plot(list(range(num_points)), x, color='cyan')
plt.plot(list(range(num_points)), x2, color='red', linestyle='solid')  # with the default solid line-style
plt.plot(list(range(num_points)), x3, color='yellow',
         linestyle='dashed')  # with a custom line-style. Other are: `dashdot` and `dotted`.
```

## Labeling Plots

#### Example with simple Title, X and Y axis labeling

```python
x = np.linspace(0, 10, 1000)

plt.plot(x, np.cos(x), color="0.75")

plt.title("A Cosine Curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

#### Advanced labeling and Legend creation

```python
num_points = 20
x = generate_increasing_data_points(high=500, low=30, n_points=num_points)
x2 = generate_decreasing_data_points(high=400, low=50, n_points=num_points)
x3 = generate_erratic_data_points(high=500, low=10, n_points=num_points)

plt.figure(figsize=(10, 7))
ax = plt.axes()
plt.plot(list(range(num_points)), x, color='cyan', label='ElectricAutomotive')
plt.plot(list(range(num_points)), x2, color='red', linestyle='solid',
         label='ShipMart')  # with the default solid line-style
plt.plot(list(range(num_points)), x3, color='yellow', linestyle='dashed',
         label='SodaPop')  # with a custom line-style. Other are: `dashdot` and `dotted`.
# Method 1: Separately define each value
# plt.xlim(-1, 20)
# plt.xticks(list(range(num_points + 1)))
# plt.title("Stock prices in the last 20 days")
# plt.xlabel("Days")
# plt.ylabel("Stock Price (LTP)")
# Method 2: Use ax.set to set all of them at once.
ax.set(
    xlim=(-1, 20), xticks=list(range(num_points + 1)),
    xlabel='Days', ylabel='Stock Price (LTP)', title='Stock prices in the last 20 days'
)

# plt.legend(loc="upper left")
# loc defines location of the legend and ncol will allow you to define number of columns. Default ncol is 1 (so they'll be stacked).
# plt.legend(loc="lower center", ncol=3)
plt.legend(loc="upper right", ncol=1, fancybox=True, framealpha=1, shadow=True, borderpad=1, frameon=True)
plt.show()
```
